```python
'''
This script reads a .csv file and generates the Compute Area Under the
Receiver Operating Characteristic Curve (ROC AUC) results from prediction scores.
'''
from classfunc import *
```

```python
# Loading and splitting the data

data_dir = 'sample_diabetes_mellitus_data.csv'
data = load_and_split(data_dir)
```

```python
# Extracting the two dataframes created from load_and_split

df_Train = data.Train
df_Test = data.Test
```

```python
# Removing NaN rows in predefined columns

nancols = ['age','gender','ethnicity']

train_clean = remove_nan(df_Train,nancols).cleaned_df
test_clean = remove_nan(df_Test,nancols).cleaned_df
```

```python
# Filling NaN with the mean value of predefined columns

meancols = ['height','weight']

train_fill = fill_nan(train_clean,meancols).df
test_fill = fill_nan(test_clean,meancols).df
```

```python
# Checking to see that NaN were replaced

train_fill[train_fill['height'].isna()]
```

```python
# Creating features: taking the log and one-hot encoding categorical variables

train_feats = log_transform(train_fill,meancols)
train_feats.apply_transform()
train_feats = train_feats.df
train_feats = one_hot_encoding(train_feats).df

test_feats = log_transform(test_fill,meancols)
test_feats.apply_transform()
test_feats = test_feats.df
test_feats = one_hot_encoding(test_feats).df
```

```python
# Checking that the data is in the format we want
train_feats.head(10)
```

```python
# Training the model on the train set
X_cols = [col for col in train_feats if col.startswith('ethnicity') or col.startswith('gender') or 
            col.startswith('hospital') or col.startswith('icu') or col.startswith('height') or 
          col.startswith('weight')]

X_cols.remove('hospital_admit_source_Chest Pain Center') # removing this dummy that does not exist in the test set

target = ['diabetes_mellitus']

mod = model(train_feats, rf, X_cols, target, hyperparams)
```

```python
# Predicting diabetes status
test_pred = mod.predict(test_feats)

test_feats['test_pred'] = test_pred.tolist()
```

```python
# Calculating ROC AUC from prediction scores
print('SCORING...')
mod.get_roc_auc_score()
```
